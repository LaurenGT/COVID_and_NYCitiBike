<font color=green><h1>Transform Data</h1></font>

```python
# drop nan from 2019

cleanMarch2019 = march2019.dropna()
cleanMarch2019.head()
```

```python
# check length compared to original march2019
print(len(march2019))
print(len(cleanMarch2019))
```

```python
# set 2019 start/end station ids to int

cleanMarch2019 = cleanMarch2019.astype({"start station id": 'int64'}, copy=False)
cleanMarch2019 = cleanMarch2019.astype({"end station id": 'int64'}, copy=False)
cleanMarch2019.head()
```

```python
# find sum of all df rows to know what the expected length will be when appending the three into one
print((len(cleanMarch2019))+(len(march2020))+(len(march2021)))
```

```python
combined2019_2020 = cleanMarch2019.append(march2020, ignore_index=True)
combineAll = combined2019_2020.append(march2021, ignore_index=True)
combineAll.head()
```

```python
# verify expected length
len(combineAll)
```

### Summary of Transformation
- dropped all NaN from 2019
- converted datatype in 2019 to match datatypes in 2020/2021
- appended cleaned files together to have one full CSV

```python
# export from notebook as new csv file to Resources folder

combineAll.to_csv("Resources/compiledMarchData.csv", index=False, encoding="utf-8")
```
